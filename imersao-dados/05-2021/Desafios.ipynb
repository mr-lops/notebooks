```python
import numpy as np


# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Aqui altera as cores do mapa de calor
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

importando dados e algumas alterações feitas em aula<br>
Desafio 04: Estudar sobre combinações de DF https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

```python
dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
dados_resultados.head()
```

```python
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0)
dados_resultados.head()
```

```python
dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa', 'ativo_moa']], on='id')
dados_combinados.head()
```

Desafio 01: Encontrar o top 10, das ações do MOA (inibidor, agonista...)

```python
TPMOA= dados_resultados.drop(['id','n_moa','ativo_moa'], axis=1).melt()
TPMOA
```

```python
plt.figure(figsize=(8, 6))
TPMOA['variable'] = TPMOA['variable'].apply(lambda x: x.split('_')[-1]) #split corta a string pelo caracter escolhido e retorna uma lista, usei o -1 para pegar o ultimo ellemento da lista
top = TPMOA['variable'].value_counts().index[0:5]
al = sns.countplot(x = 'variable', data=TPMOA.query('variable in @top'), order = top)
al.set_title('Top 10 ações do MOA', fontsize=25)
plt.ylabel("Frequencia", fontsize=15)
plt.xlabel("Ação MOA", fontsize=15)
```

Desafio 02: Criar a coluna eh_controle para quando na linha tratamento == com_controle

```python
dados_combinados['eh_controle'] = (dados_combinados['tratamento'] == 'com_controle')
dados_combinados.head()
```

Desafio 03: Criar três colunas para indicar o tempo 24, 48, 72

```python
dados_combinados['24'] = (dados['tempo'] == 24)
dados_combinados['48'] = (dados['tempo'] == 48)
dados_combinados['72'] = (dados['tempo'] == 72)
dados_combinados.head()
```

Desafio 05: Fazer análise mais detalhada considerando tempo e dose, para comparar as distribuições (Escolher uma droga e comparar com controle)

```python
composto_principal = dados_combinados['composto'].value_counts().index[:2]
plt.figure(figsize=(12,8))
sns.catplot(x="composto", y="g0", hue="tempo",
            col="dose", kind= 'box', data=dados_combinados.query('composto in @composto_principal'))
```

Desafio 06: Descobrir se tem algum composto que dependendo da configuração do experimento, ativa ou não ativa algum MOA <br>
Desafio 07: Descobrir se tem algum composto que dependendo da configuração do experimento, ativa MOAs diferentes

```python
top5 = dados_combinados[['composto','ativo_moa','n_moa','dose','eh_controle','24','48','72']].value_counts().reset_index().drop_duplicates(subset='composto', keep='first').reset_index(drop = True)[:5]

top5.columns = ['Composto', 'Ativou MOA?','N.º MOA Ativou', "Tipo de Dose", "É Controle?", "Exposto a 24hrs?","Exposto a 48hrs?","Exposto a 72hrs?", "Frequencia"]
top5
```

Desafio 08: Resumo do que você aprendeu com os dados

    Observamos na tabela acima que algumas configurações acontecem com mais frequencia em relação a cada composto, principalmente envolvendo a Dose numero 1 e o tempo se exposição de 48 horas.
